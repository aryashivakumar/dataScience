**Task5**:

Create and run a stright line linear regression model on the howell data set with height predicting weight. Visualize your model and its posterior.

```python
with pm.Model() as model_lb:

    #priors for the linear part of our model
    Î± = pm.Normal("Î±", mu=0, sigma=100)
    Î² = pm.Normal("Î²", mu=0, sigma=10)

    #this is the linear part of our model
    Î¼ = pm.Deterministic("Î¼", Î± + Î² * howell.height)

    #prior for the standard deviation of our likelihood
    #Cauchy is a T dist with nu = 1
    Ïƒ = pm.HalfCauchy("Ïƒ", 10)

    #likelihood
    y_pred = pm.Normal("y_pred", mu=Î¼, sigma=Ïƒ, observed=howell.weight)

    #inference data object
    idata_lb = pm.sample()
```

```python
posterior = az.extract(idata_lb, num_samples=100)

# grabbing x values for graphing.
x_plot = xr.DataArray(
    np.linspace(howell.height.min(), howell.height.max(), 50),
    dims="plot_id"
    )

# this creates the expected line, the path we predict temperature and
#rented bike count deviates from
mean_line = posterior["Î±"].mean() + posterior["Î²"].mean() * x_plot

#These are a 50 lines that our model came up with
lines = posterior["Î±"] + posterior["Î²"] * x_plot

#set up plot
_, ax = plt.subplots()

#plots 50 other lines our model came up with.
ax.plot(x_plot, lines.T, c="C1", alpha=0.2, label="lines")

#plots the mean line
ax.plot(x_plot, mean_line, c="C0", label="mean line")

#plot the raw data
ax.plot(howell.height, howell.weight, "C2.", zorder=-3)

#label axes and create legend
ax.set_xlabel("height")
ax.set_ylabel("weight")
```

**Task7**:

Create and visualize the posterior of a generalized linear model that improves the on the fit of the model you made in Task5.

```python
# Scale height for better sampling properties
height_scaled = (howell.height - howell.height.mean()) / howell.height.std()

with pm.Model() as model_exp_fixed:
    # Tighter priors for scaled predictor
    alpha = pm.Normal("alpha", mu=0, sigma=1)
    beta = pm.Normal("beta", mu=0, sigma=1)

    # Exponential mean model
    mu = pm.Deterministic("mu", pm.math.exp(alpha + beta * height_scaled))

    # Prior for the standard deviation of our likelihood
    # Using HalfNormal with a reasonable sigma
    sigma = pm.HalfNormal("sigma", sigma=5) # Adjust sigma prior if needed, e.g., based on residuals

    # Likelihood
    y_pred = pm.Normal("y_pred", mu=mu, sigma=sigma, observed=howell.weight)

    # Inference data object and posterior predictive sampling
    idata_exp_fixed = pm.sample(target_accept=0.9) # Increase target_accept to mitigate divergences
    pm.sample_posterior_predictive(idata_exp_fixed, model=model_exp_fixed, extend_inferencedata=True)
```

```python
# Now, for plotting:
posterior = az.extract(idata_exp_fixed, num_samples=100)

# Create x values for plotting, in original scale
x_plot_original = np.linspace(howell.height.min(), howell.height.max(), 100)
# Scale these x values for prediction
x_plot_scaled = (x_plot_original - howell.height.mean()) / howell.height.std()

# Calculate the mean line using the extracted posteriors and scaled x_plot
# .item() to get scalar from xarray DataArray for mean_alpha and mean_beta
mean_alpha = posterior["alpha"].mean().item()
mean_beta = posterior["beta"].mean().item()
mean_line_predicted = np.exp(mean_alpha + mean_beta * x_plot_scaled)

# For HDI, we need y_pred_q from posterior predictive.
y_pred_q = idata_exp_fixed.posterior_predictive["y_pred"].quantile(
    [0.03, 0.97, 0.25, 0.75], dim=["chain", "draw"]
)

# Create a temporary DataFrame to process height and y_pred_q values
temp_df = pd.DataFrame({'height': howell.height.values})
for i in range(4):
    temp_df[f'y_pred_q_{i}'] = y_pred_q[i].values

# Group by height and take the mean of the y_pred_q values for duplicate heights
# Then sort by height to ensure strictly increasing x for interpolator
df_for_interp = temp_df.groupby('height').mean().sort_values(by='height').reset_index()

y_hat_bounds_interpolators = []
for i in range(4):
    # Use the unique sorted heights and their corresponding y_pred_q values
    interpolator = PchipInterpolator(
        df_for_interp['height'].values,
        df_for_interp[f'y_pred_q_{i}'].values
    )
    y_hat_bounds_interpolators.append(interpolator(x_plot_original))

# Set up plot
_, ax = plt.subplots(figsize=(10, 6))

# Plots raw data
ax.plot(howell.height, howell.weight, "C2.", zorder=-3, label="Observed Data")

# Plots the mean line
ax.plot(x_plot_original, mean_line_predicted, c="C0", label="Mean Prediction")

# Graphs the 94% and 50% HDIs
ax.fill_between(x_plot_original, y_hat_bounds_interpolators[0], y_hat_bounds_interpolators[1], color="C1", alpha=0.2, label="94% HDI")
ax.fill_between(x_plot_original, y_hat_bounds_interpolators[2], y_hat_bounds_interpolators[3], color="C1", alpha=0.4, label="50% HDI")

# Labels and legend
ax.set_xlabel("Height (cm)")
ax.set_ylabel("Weight (kg)")
ax.set_title("Generalized Linear Model (Exponential Mean) for Howell Data")
plt.legend()
plt.show()
```
